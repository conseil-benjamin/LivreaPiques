```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import plotly.express as px

## Load the data
df = pd.read_csv("CleanedAuthors.csv")

display(df.head())
```

```python
## restreindre sur birthplace et gender
df = df[['birthplace', 'genre_1']]
display(df.head())
```

```python
# tableau de contingence
contingency_table = pd.crosstab(df['birthplace'], df['genre_1'])
display(contingency_table)

## enlever les ligne qui ne sont pas que des lettre
contingency_table = contingency_table[contingency_table.index.str.isalpha()]
contingency_table = contingency_table[contingency_table.columns[contingency_table.columns.str.isalpha()]]
display(contingency_table)
```

```python
import pandas as pd
from sqlalchemy import create_engine, select, Table, MetaData
from sqlalchemy.orm import sessionmaker

# Connexion à la base de données
engine = create_engine(database_url)
Session = sessionmaker(bind=engine)
session = Session()

# Créer l'objet MetaData sans `bind`
metadata = MetaData()

# Charger les tables book et awards en utilisant `autoload_with=engine`
book_table = Table('book', metadata, autoload_with=engine)
series_table = Table('series', metadata, autoload_with=engine)

# Requête pour obtenir les IDs de livres et awards selon leurs noms
def get_id_by_name(table, name_column, name_value, search_column='id'):
    result = session.execute(select(table.c[search_column]).where(table.c[name_column] == name_value)).first()
    return result[0] if result else None

# Parcours du DataFrame pour récupérer les IDs et afficher un message après chaque insertion dans `associations`
associations = []
for index, row in data2.iterrows():
    book_id = get_id_by_name(book_table, 'title', row['title'], search_column='id_book')
    series_id = get_id_by_name(series_table, 'name_series', row['series_name'], search_column='id_series')
    
    # Vérifie que les deux IDs existent avant de les insérer
    if book_id and series_id:
        associations.append({'id_book': book_id, 'id_series': series_id})
        print(f"Association ajoutée : livre ID {book_id}, series ID {series_id}")

# sauvegarder les associations dans un fichier csv
df_associations = pd.DataFrame(associations)
df_associations.to_csv('associationsSeries.csv', index=False)


# Charger les associations depuis le fichier CSV
associations = pd.read_csv('associations.csv')

# renommer les colonnes
associations = associations.rename(columns={'book_id': 'id_book', 'series_id': 'id_series'})

# enlever les doublons de (id_book, id_award)
associations = associations.drop_duplicates()

display(associations.head())

# Insérer les associations dans la table SQL 'book_award'
associations.to_sql('book_series', con=engine, if_exists='append', index=False)


# Ferme la session après insertion
session.close()
```

```python
import pandas as pd
import re
import ftfy

# Fonction pour détecter des caractères non-ASCII
def contains_non_ascii(text):
    return bool(re.search(r'[^\x00-\x7F]', str(text)))

# Fonction pour ré-encoder les titres en UTF-8
def reencode_to_utf8(text):
    try:
        # Essayer de décoder en UTF-8 directement
        return text.encode('latin1').decode('utf-8')
    except (UnicodeEncodeError, UnicodeDecodeError):
        try:
            # Essayer de décoder en UTF-8 après avoir encodé en latin1
            return text.encode('utf-8').decode('latin1')
        except (UnicodeEncodeError, UnicodeDecodeError):
            return text

# Fonction pour corriger les erreurs d'encodage avec ftfy
def fix_encoding(text):
    return ftfy.fix_text(text)

# Lire le fichier CSV
df = pd.read_csv('bigboss_book.csv')

# Colonnes à traiter
columns_to_fix = ['title', 'publisher', 'series', 'author', 'original_title', 'genre_and_votes', 'settings', 'characters', 'awards', 'description']

# Appliquer les fonctions sur les colonnes spécifiées
for column in columns_to_fix:
    df_with_non_ascii = df[df[column].apply(contains_non_ascii)]
    df_with_non_ascii[column] = df_with_non_ascii[column].apply(reencode_to_utf8)
    df_with_non_ascii[column] = df_with_non_ascii[column].apply(fix_encoding)
    df.update(df_with_non_ascii)

# Sauvegarder le fichier CSV corrigé
df.to_csv('books_corrected.csv', index=False)

# Afficher les lignes contenant des caractères non-ASCII pour chaque colonne
for column in columns_to_fix:
    print(f"Non-ASCII characters in column '{column}':")
    print(df[df[column].apply(contains_non_ascii)][column])
```

```python
#faire une acp sur tout les variable quantitative de df3
import matplotlib.pyplot as plt
import seaborn as sns

from sklearn.decomposition import PCA
from sklearn.preprocessing import StandardScaler

df4 = df3.select_dtypes(include=['float64', 'int64'])
df4 = df4.dropna()
print(len(df4))
print(df4.columns)
scaler = StandardScaler()
df4_scaled = scaler.fit_transform(df4)
pca = PCA()
pca.fit(df4_scaled)
pca_res = pca.transform(df4_scaled)

```

```python
#Les valeurs propres des composantes : pca.singular values
print("Singular values")
singVal = pca.singular_values_
print(singVal)
print("-"*30)
#Les pourcentages de valeurs propres : pca.explained variance ratio
print("Explained variance ratio")
VarRatio = pca.explained_variance_ratio_
print(VarRatio)
print("-"*30)
#Les valeurs propres : pca.explained variance
print("cumulative explained variance")
CumVar = np.cumsum(pca.explained_variance_ratio_)
print(CumVar)
print("-"*30)
```

```python
##aficher nuage de points
plt.scatter(pca_res[:,0], pca_res[:,1])
plt.title("PCA")
plt.xlabel("Dim1")
plt.ylabel("Dim2")
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import re

# Fonction pour extraire le genre avec le plus de votes
def get_top_genre(genre_and_votes):
    if pd.isna(genre_and_votes):
        return None
    # Séparer les paires "genre nombre_de_votes"
    genre_votes = re.findall(r'(\D+)\s(\d+)', genre_and_votes)
    if not genre_votes:
        return None
    # Trouver le genre avec le plus de votes
    top_genre = max(genre_votes, key=lambda x: int(x[1]))[0].strip()
    return top_genre

# Fonction pour regrouper les genres similaires
def group_genres(genre):
    genre = genre.lower()
    if 'cultural' in genre:
        return 'Cultural'
    elif 'christian' in genre:
        return 'Christian'
    elif 'historical' in genre:
        return 'Historical'
    elif 'romance' in genre:
        return 'Romance'
    elif 'thriller' in genre:
        return 'Thriller'
    elif 'fiction' in genre:
        return 'Fiction'
    elif 'mystery' in genre:
        return 'Mystery'
    elif 'fantasy' in genre:
        return 'Fantasy'
    elif 'philosophy' in genre:
        return 'Philosophy'

# Lire le fichier CSV nettoyé
df = pd.read_csv('books_corrected.csv')

# Extraire l'année de publication
df['date_published'] = pd.to_datetime(df['date_published'], errors='coerce').dt.year

# Supprimer les lignes avec des années de publication manquantes
df_cleaned = df.dropna(subset=['date_published', 'genre_and_votes'])

# Appliquer la fonction pour extraire le genre dominant
df_cleaned['top_genre'] = df_cleaned['genre_and_votes'].apply(get_top_genre)

# Supprimer les lignes où le genre dominant n'a pas pu être déterminé
df_cleaned = df_cleaned.dropna(subset=['top_genre'])

# Regrouper les genres similaires
df_cleaned['grouped_genre'] = df_cleaned['top_genre'].apply(group_genres)

# Agréger les données par année et genre regroupé
df_genre_year = df_cleaned.groupby(['date_published', 'grouped_genre']).size().reset_index(name='book_count')

# Pivot pour préparer les données à tracer
df_pivot = df_genre_year.pivot(index='date_published', columns='grouped_genre', values='book_count').fillna(0)

# Tracer un graphique en ligne
plt.figure(figsize=(12, 8))
df_pivot.plot(kind='line', figsize=(14, 8))

# Ajouter les étiquettes et le titre
plt.title('Nombre de livres par genre regroupé en fonction des années de publication', fontsize=16)
plt.xlabel('Année de publication', fontsize=12)
plt.ylabel('Nombre de livres', fontsize=12)

# Afficher la légende et le graphique
plt.legend(title='Genres')
plt.grid(True)
plt.tight_layout()
plt.show()
```
